````python
import pandas as pd
import numpy as np

DATA = "../data"

results = pd.read_csv(f"{DATA}/results.csv")
train   = pd.read_csv(f"{DATA}/wc_teams_train.csv")
test    = pd.read_csv(f"{DATA}/wc_teams_test.csv")
elo_wc  = pd.read_csv(f"{DATA}/elo_ratings_wc2026.csv")
fix     = pd.read_csv(f"{DATA}/wc2026_fixtures.csv")

print("Fichiers chargés")
````

## Étape 1 — Exploration fichier par fichier

````python
def explore(name, df):
    print(f"\n{'='*60}")
    print(f"  {name}")
    print(f"{'='*60}")
    print(f"Shape        : {df.shape[0]:,} lignes x {df.shape[1]} colonnes")
    print(f"Colonnes     : {list(df.columns)}")
    display(df.head(3))
    missing = df.isnull().sum()
    missing = missing[missing > 0]
    if missing.empty:
        print("Valeurs manquantes : aucune")
    else:
        print("Valeurs manquantes :")
        for col, n in missing.items():
            print(f"  {col}: {n} ({n/len(df)*100:.1f}%)")

explore("results.csv", results)
explore("wc_teams_train.csv", train)
explore("wc_teams_test.csv", test)
explore("elo_ratings_wc2026.csv", elo_wc)
explore("wc2026_fixtures.csv", fix)
````

## Étape 2 — Clés de jointure et standardisation des noms d'équipes

````python
# Noms d'équipes par source
teams_test    = set(test["team"].unique())
teams_train   = set(train["team"].unique())
teams_elo     = set(elo_wc["country"].unique())
teams_fix_gs  = set(fix[fix["stage"]=="Group Stage"]["home_team"]) | \
                set(fix[fix["stage"]=="Group Stage"]["away_team"])
teams_results = set(results["home_team"]) | set(results["away_team"])

print(f"wc_teams_test    : {len(teams_test)} équipes")
print(f"wc_teams_train   : {len(teams_train)} équipes")
print(f"elo_ratings      : {len(teams_elo)} pays")
print(f"fixtures GS      : {len(teams_fix_gs)} équipes")
print(f"results.csv      : {len(teams_results)} équipes")

print(f"\nEditions dans wc_teams_train : {sorted(train['version'].unique())}")
print(f"Années dans elo_ratings       : {elo_wc['year'].min()} → {elo_wc['year'].max()}")
````

````python
# Divergences de noms : wc_teams_test vs wc2026_fixtures
print("=== test vs fixtures (Group Stage) ===")
print(f"Dans test mais PAS dans fixtures : {sorted(teams_test - teams_fix_gs)}")
print(f"Dans fixtures mais PAS dans test : {sorted(teams_fix_gs - teams_test)}")

print("\n=== test vs elo_ratings (country) ===")
print(f"Dans test mais PAS dans elo : {sorted(teams_test - teams_elo)}")

print("\n=== test vs results.csv ===")
print(f"Dans test mais PAS dans results : {sorted(teams_test - teams_results)}")
````

````python
# Mapping nécessaire : test → fixtures
# Ces noms doivent être alignés avant toute jointure
NAME_MAP_TEST_TO_FIXTURE = {
    "United States": "USA",
    "South Korea":   "Korea Republic",
    "Iran":          "IR Iran",
    "Cape Verde":    "Cabo Verde",
    "DR Congo":      "Congo DR",
    "Ivory Coast":   "Côte d'Ivoire",
    "Czech Republic":"Czechia",
    "Turkey":        "Türkiye",
    # Curaçao à vérifier (caractère spécial)
}
print("Mapping défini. Valider que les 48 équipes de test couvrent bien les 48 de fixtures après mapping.")

test_mapped = test["team"].map(lambda x: NAME_MAP_TEST_TO_FIXTURE.get(x, x))
coverage = teams_fix_gs - set(test_mapped)
print(f"Non couverts après mapping : {sorted(coverage)}")
````

## Étape 4 bis — Architecture proposée

### Problème fondamental de granularité

```
results.csv     : 1 ligne = 1 match (home vs away)
wc_teams_*      : 1 ligne = 1 équipe × 1 édition CdM
elo_ratings     : 1 ligne = 1 équipe × 1 snapshot annuel
wc2026_fixtures : 1 ligne = 1 match 2026
```

Le modèle prédit H/D/A pour un match → features **par match**. Les wc_teams doivent être jointes **deux fois** par match (home + away).

### Architecture propre

```
SOURCE A — results.csv (pipeline existant)
  → ELO pré-match, forme 5 matchs, H2H
  → leakage-safe avec shift(1)
  → 49 215 matchs toutes compétitions

SOURCE B — wc_teams_train.csv
  → Filtrer : WC uniquement dans results.csv
  → Extraire l'année du match (ex: WC 2022 = 2022)
  → Join sur (home_team, year) → home_fifa_rank, home_squad_age...
  → Join sur (away_team, year) → away_fifa_rank, away_squad_age...
  → Créer diff_X = home_X - away_X pour chaque feature
  → ATTENTION : utiliser version = année du tournoi, pas la date du match

SOURCE C — elo_ratings_wc2026.csv
  → Apporte : wins/losses/draws/goals cumulés, rank_avg, matches_total
  → Join sur (country, year) avec snapshot_date < date_début_tournoi
  → Alternative : déjà couvert par elo_history.csv pour l'ELO pur
  → Valeur ajoutée réelle : stats cumulées (win rate, goal diff cumulé)

PRÉDICTION 2026 :
  → wc_teams_test.csv joint sur wc2026_fixtures.csv
  → JAMAIS inclus dans l'entraînement
  → Mapping noms : test → fixtures (9 divergences identifiées)
```

### Features candidates prioritaires

| Feature | Source | Utilité | Priorité |
|---|---|---|---|
| `fifa_rank` | wc_teams | Signal de force global officiel | ⭐⭐⭐ |
| `fifa_points` | wc_teams | Continu, plus précis que le rang | ⭐⭐⭐ |
| `world_cup_participations_before` | wc_teams | Expérience CdM | ⭐⭐ |
| `groups_passed_before / round16_before` | wc_teams | Historique CdM | ⭐⭐ |
| `goals_scored_last_4y` | wc_teams | Forme sur 4 ans (mais fenêtre longue) | ⭐ |
| `squad_avg_age` | wc_teams | Maturité vs expérience | ⭐ |
| `is_host` | wc_teams | Avantage terrain | ⭐⭐ |
| `win_rate_cumul` | elo_ratings | Calculé : wins/(wins+losses+draws) | ⭐⭐ |
| `goal_diff_cumul` | elo_ratings | Calculé : goals_for - goals_against | ⭐ |
